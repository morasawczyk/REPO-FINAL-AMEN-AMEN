## Definiciones basicas de programacion

Definamos algunos conceptos que forman parte del punto de partida y a lo largo del curso profundizaremos en algunos de ellos.

### Algoritmos

Un algoritmo se define como una **descripción no ambigua y precisa de las acciones** que hay que realizar para resolver un problema bien definido en un tiempo finito. 

### Paradigma de programacion

Un paradigma de programación es una **propuesta tecnológica adoptada por una comunidad de programadores y desarrolladores**. Representa un **enfoque particular o filosófico** para diseñar soluciones. Los paradigmas difieren unos de otros, en los conceptos y la forma de **abstraer** los elementos involucrados en un problema, así como en los pasos que integran la solución del problema. En general la mayoría son variantes de los dos tipos principales, **imperativa** y **declarativa**:

* **Programación imperativa**: es como dar **instrucciones paso a paso**, similar a una receta de cocina. Le decis a la computadora exactamente *qué hacer* y *cómo hacerlo*. Por ejemplo, "toma este número, súmale 5, guárdalo acá, mostralo en pantalla". Pensa en esto como darle indicaciones detalladas a alguien para llegar a un lugar. Lenguajes como C, BASIC y Python (en su uso más común) siguen este enfoque.

* **Programación declarativa**: en lugar de explicar cada paso, **describis lo que queres obtener** y dejas que la computadora resuelva cómo lograrlo. Es como decirle a un taxi "llévame al aeropuerto" sin especificar la ruta. Incluye dos estilos principales:
  - **Lenguajes funcionales** (como Haskell): se centran en aplicar funciones a los datos
  - **Lenguajes lógicos** (como Prolog): definen hechos y reglas, y la computadora encuentra soluciones que cumplan esas condiciones

### Python

Python es un **lenguaje de programación versátil, potente y accesible** que se ha convertido en uno de los más populares del mundo. Desde su creación a finales de los ochenta por Guido van Rossum, Python ha evolucionado hasta su versión 3.12/3.13 (2025), manteniendo su filosofía original de simplicidad y legibilidad.

Su **sintaxis clara y concisa** permite expresar conceptos en menos líneas de código que otros lenguajes, mientras que sus **múltiples paradigmas de programación** (orientado a objetos, imperativo, funcional) ofrecen flexibilidad para abordar problemas de diferentes maneras. Python combina un **sistema de tipado dinámico** con verificación de tipos opcional mediante "type hints", proporcionando lo mejor de ambos mundos: flexibilidad durante el desarrollo y seguridad en proyectos complejos.

![Imagen de Guido van Rossum](../img/guidoVanRossum.jpg)

> Curiosamente, el apellido de Guido coincide con el nombre de la compañía en la obra teatral de ciencia ficción **[R.U.R - Rossum's Universal Robots](https://es.wikipedia.org/wiki/R.U.R._(Robots_Universales_Rossum))** (1921), que introdujo la palabra "robot" al vocabulario moderno. La palabra fue acuñada por Josef Čapek a partir del término checo **robota** (trabajo), estableciendo una interesante conexión simbólica entre el creador de Python y los orígenes de la robótica en la ficción.


![](../img/rur.jpg)

Entre las características que han impulsado a Python como lenguaje dominante en ciencia de datos, IA y procesamiento de señales se encuentran:

* **Ecosistema científico maduro**: Librerías como NumPy, SciPy, y Matplotlib son estándares de facto para procesamiento numérico, científico y visualización de datos.
* **Arquitectura modular**: Permite dividir programas en módulos reutilizables, facilitando el desarrollo colaborativo y el mantenimiento.
* **Interpretación y compilación JIT**: Python combina la flexibilidad de un lenguaje interpretado con optimizaciones avanzadas para rendimiento.
* **Comunidad activa**: Con más de 500,000 paquetes disponibles en PyPI (2025) y una extensa documentación mantenida por la comunidad.
* **Especialización en audio**: Bibliotecas como Librosa, PyDub y Pedalboard han convertido a Python en el lenguaje preferido para procesamiento de audio digital.

El lenguaje es de **código abierto** y está disponible gratuitamente para todas las plataformas desde el [sitio web oficial de Python](https://www.python.org/), administrado por la Python Software Foundation.

> El nombre "Python" proviene del grupo cómico británico "["Monty Python's Flying Circus"](https://es.wikipedia.org/wiki/Monty_Python%27s_Flying_Circus)" y no tiene relación con las serpientes, aunque la comunidad ha adoptado este animal como mascota no oficial.

* Python te permite **separar el programa en módulos** que pueden reusarse en otros programas en Python.
* Python es un **lenguaje interpretado**, lo cual puede ahorrarte mucho tiempo durante el desarrollo ya que no es necesario compilar.
* Python permite **escribir programas compactos y legibles**.

El intérprete de Python y la extensa biblioteca estándar están a libre disposición en forma binaria y el código fuente para las principales plataformas desde el [sitio web de Python](https://www.python.org/), y puede distribuirse libremente [G. Van Rossum., 2017]. 

### Caracteristicas de Python

* **Libre y open source**. Python es estable, confiable, con gran soporte y una gran comunidad de desarrolladores.
* **Propósito general**. Python viene con una gran cantidad de librerías instaladas, haciéndolo muy versátil. Se puede usar directamente en el shell (microtasks), o de forma interactiva, o en scripts, o para crear software comercial con GUI.
* **Cross platform**. Puede ejecutarse código de Python en cualquier arquitectura, versiones para Windows, Linux/Unix, MAC OS X y muchos otros.
* **Objetos**. Los tipos son asociados con objetos, no con variables. Python opera generalmente en un alto nivel de abstracción. Las reglas de sintaxis son muy simples.
* **Expresivo**. Las buenas prácticas de programación, "Zen de Python", además de una sintaxis muy simple hacen un codigo muy facil de leer en menos líneas, que en otros lenguajes.
* **Multiparadigma**. Si bien predomina el POO, es posible trabajar con otros paradigmas.
* **Código eficiente**. Los módulos numéricos de Python son computacionalmente eficientes.
